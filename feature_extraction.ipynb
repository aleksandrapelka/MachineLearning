## <a name='0'></a> 1. Import bibliotek

```python
import pandas as pd
import numpy as np
import sklearn
import pandas_datareader.data as web
```

## <a name='1'></a> 2. Załadowanie danych (notowania spółki Amazon)

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()    
```

## <a name='2'></a> 3. Utworzenie kopii danych

```python
df = df_raw.copy()
df = df[:5]
df.info()
```

## <a name='3'></a> 4. Generowanie nowych zmiennych

```python
df.index
```

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

## <a name='4'></a> 5. Dyskretyzacja zmiennej ciągłej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

#### Podział zmiennej ciągłej na 3 kategorie

```python
df['height_cat'] = pd.cut(x=df.height, bins=3)
df
```

#### Utworzenie własnych przedziałów

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195))
df
```

#### Utworzenie własnych etykiet zamiast przedziałów wartości

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195), labels=['small', 'medium', 'high'])
df
```

#### Utworzenie zmiennych binarnych

```python
pd.get_dummies(df, drop_first=True, prefix='height')
```

## <a name='5'></a> 6. Ekstrakcja cech

```python
df = pd.DataFrame(data={'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

#### Utworzenie nowej kolumny przechowującej liczbę języków

```python
df['lang_number'] = df['lang'].apply(len)
df
```

#### Utworzenie nowej kolumny przechowującej informację o obecności języka polskiego

```python
df['PL_flag'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0)
df
```
